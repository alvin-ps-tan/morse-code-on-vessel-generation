```python
import cv2
import glob 

import numpy as np  
```

```python
SOURCE_FOLDER = "raw_data/26_evening/"
SOURCE_VIDEOS = glob.glob(SOURCE_FOLDER + "*.mp4")
```

```python
video = cv2.VideoCapture(SOURCE_VIDEOS[0])
fps = video.get(cv2.CAP_PROP_FPS)
width  = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(SOURCE_FOLDER + "output.mp4", fourcc, fps, (width, height))
```

```python
def write_frames(cap):

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        out.write(frame)
```

```python
video1 = cv2.VideoCapture(SOURCE_VIDEOS[0])
video2 = cv2.VideoCapture(SOURCE_VIDEOS[1])

write_frames(video1)
write_frames(video2)

video1.release()
video2.release()
out.release()
```
